```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as rd
```

```python
m,b = 0.75, 0.25

X = np.arange(0,100,1)
y = m*X + b
```

```python
for i in range(np.size(y)):
    y[i] += rd.randint(-40,40)
    
```

```python
plt.plot(X,y)
```

```python
plt.scatter(X,y)
```

```python
mean_X = np.mean(X)
mean_y = np.mean(y)
mean_X,mean_y
```

```python
n = len(X)
n
```

```python
y[:3]
```

```python
numer = 0
denom = 0
m_graph = {}
b_graph = {}
error = {}
for i in range(n):
    numer += (X[i] - mean_X) * (y[i] - mean_y)
    denom += (X[i] - mean_X)**2
    m_graph[i] = numer / denom
    b_graph[i] = mean_y - (m_graph[i] * mean_X)
    error[i] = sum((y[:i] -(m_graph[i] * X[:i] + b))**2)
    
    
m_pred = numer / denom
b_pred = mean_y - (m_pred * mean_X)
m_graph, b_graph
```

```python
plt.plot(m_graph.keys(), m_graph.values())
min(m_graph.values())
```

```python
plt.plot(b_graph.keys(), b_graph.values())
```

```python
plt.scatter(X,y)
plt.plot(X, m_pred*X + b_pred, 'y')
```
